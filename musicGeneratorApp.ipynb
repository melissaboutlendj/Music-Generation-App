```python
from diffusers.pipelines.audioldm2 import AudioLDM2Pipeline
```

```python
import torch
```

```python
import torch

print(torch.cuda.is_available())

model_id = "cvssp/audioldm2"
pipe = AudioLDM2Pipeline.from_pretrained(
model_id,
    torch_dtype = torch.float32
    # move the modele to the GPU
).to("cpu")
```

```python
prompt = "A cheerful ukulele strumming in a beachside jam."
```

```python
generator = torch.Generator().manual_seed(0)
```

```python
audio = pipe(prompt,
             audio_length_in_s = 10.24,
             generator = generator,
             ).audios[0]
```

```python
from IPython.display import Audio
Audio(audio, rate = 16000 )
# we will listen to our first music
```

```python
import gradio as gr
```

```python
# create a function to generate the audio
def create_music(prompt):
    negative_prompt = "Low quality"
    # let's paste our pipeline
    audio = pipe(prompt,
             negative_prompt = negative_prompt,
             audio_length_in_s = 10.24,
             generator = generator,
             ).audios[0]
    return 16000, audio
```

```python
# create our interface
interface = gr.Interface(
    title = "Music Generation App",
    examples = ["A serene and elegant piano composition, evoking a sense of royal grandeur. The music flows with graceful, melancholic melodies, featuring delicate arpeggios and lush harmonic progressions. It carries a contemplative, introspective tone, reminiscent of a regal setting, with slow, deliberate tempo changes. The experience is both uplifting and introspective, with soft, cascading piano notes and rich, resonant chords, painting a picture of a majestic, peaceful palace garden at sunset. A touch of modern minimalism intertwined with classical beauty"],
    fn = create_music,
    inputs = gr.Textbox(),
    outputs = "audio",
).launch(debug = True, share= True)
```
